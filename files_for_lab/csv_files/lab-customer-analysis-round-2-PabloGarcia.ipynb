```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('marketing_customer_analysis.csv')
```

```python
#1.Show the dataframe shape.
df.shape
```

```python
#2.Standardize header names
df.columns
```

```python
df = df.rename({"Unnamed: 0": 'Index', 'EmploymentStatus': 'Employment Status'}, axis=1)
```

```python
# Change columns to lowercase
cols1 = []
for value in df.columns:
    cols1.append(value.lower())
df.columns = cols1
```

```python
df
```

```python
#3.Which columns are numerical?
numeric_df = df._get_numeric_data()
numeric_df
```

```python
#4. Which columns are categorical?
categorical_df = df.select_dtypes('object')
categorical_df
```

```python
#5.Check and deal with NaN values.
# Percentage of NaN values in each column
null_df = pd.DataFrame(df.isna().sum()/len(df)*100)
null_df
```

```python
# Drop "vehicle type" column due to high number of NaN values.

df = df.drop(['vehicle type'], axis=1)
```

```python
df
```

```python
df.dtypes
```

```python
# If dtype is float64 > replace NaN values with average.
```

```python
months_since_mean = np.mean(df["months since last claim"])
df['months since last claim'] = df['months since last claim'].fillna(months_since_mean)
```

```python
number_of_open_mean = np.mean(df["number of open complaints"])
df["number of open complaints"] = df["number of open complaints"].fillna(number_of_open_mean)
```

```python
# If dtype is object > replace NaN values with "Unknown".
df['state'] = df['state'].fillna("Unknown")
df['response'] = df['response'].fillna("Unknown")
df['vehicle class'] = df['vehicle class'].fillna("Unknown")
df['vehicle size'] = df['vehicle size'].fillna("Unknown")
```

```python
df
```

```python
#Confirm that we do not have any NaN values in the dataframs

null_df = pd.DataFrame(df.isna().sum()/len(df)*100)
null_df
```

```python
#6. Datetime format - Extract the months from the dataset and store in a separate column. 
```

```python
# Convert the 'effective to date' column to DateTime format
df['effective to date'] = pd.to_datetime(df['effective to date'], errors='coerce')
```

```python
df.dtypes
```

```python
# Try to use lambda function
df['month'] = pd.DatetimeIndex(df['effective to date']).month
```

```python
df.head()
```

```python
#Then filter the data to show only the information for the first quarter , ie. January, February and March. 
#Hint: If data from March does not exist, consider only January and February.
df_first_quarter = df[(df.month == 1) | (df.month == 2) | (df.month == 3)]
```
